```python
import pandas as pd
data= pd.read_csv("sephora_skincare_reviews.csv",index_col=False,keep_default_na=True)
```

# Analyse de la fréquence des mots

```python
from collections import Counter
import nltk
from nltk.tokenize import word_tokenize

opinions = data['feedback']

all_opinions = ' '.join(opinions.dropna())
tokens = word_tokenize(all_opinions)

word_freq = Counter(tokens)

print(word_freq.most_common(10))
```

# Nettoyage des données

```python
from string import punctuation
nltk.download('stopwords')
cleaned_text = ' '.join([word.lower() for word in tokens if word.isalnum() and word not in punctuation])

cleaned_word_freq = Counter(word_tokenize(cleaned_text))

print(cleaned_word_freq.most_common(10))
```

# Traduisez chaque élément en anglais

```python
from googletrans import Translator
import pandas as pd
from requests.exceptions import ReadTimeout
import time

translator = Translator()

text_to_translate_series = data['opinion']

def translate_with_retry(text, max_retries=3):
    retries = 0
    while retries < max_retries:
        try:
            return translator.translate(text, dest='en').text
        except ReadTimeout:
            print(f"Translation timed out. Retrying ({retries + 1}/{max_retries}) after a short delay...")
            time.sleep(1)  
            retries += 1
    print(f"Translation failed after {max_retries} retries. Returning an empty string.")
    return ""

translated_series = text_to_translate_series.apply(translate_with_retry)

print(translated_series)
```

```python
data['opinion'] = translated_series
```

# Afficher les mots les plus fréquents
 

```python
from textblob import TextBlob
from textblob_fr import PatternTagger, PatternAnalyzer
from langdetect import detect, LangDetectException

def get_sentiment(opinion):
    try:
        language = detect(opinion)
        
        if language == 'fr':
            analysis = TextBlob(str(opinion), pos_tagger=PatternTagger(), analyzer=PatternAnalyzer())
            return analysis.sentiment[0]  
        elif language == 'en':
            analysis = TextBlob(str(opinion))
            return analysis.sentiment.polarity
        else:
            return 0  
    except LangDetectException:
        return 0  

data['sentiment'] = opinions.apply(get_sentiment)

def classify_sentiment(polarity):
    if polarity > 0:
        return 'bon'
    elif polarity < 0:
        return 'mauvais'
    else:
        return 'neutre'

data['classification'] = data['sentiment'].apply(classify_sentiment)

print(data)
print(data[['opinion', 'sentiment', 'classification']])

```
